# Sobre os dados

O Fundo de Financiamento Estudantil (FIES) é um programa do Ministério da Educação destinado a financiar a graduação na educação superior de estudantes matriculados em cursos superiores não gratuitas na forma da Lei 10.260/2001. Podem recorrer ao financiamento os estudantes matriculados em cursos superiores que tenham avaliação positiva nos processos conduzidos pelo Ministério da Educação.

Os dados são referentes ao FIES de 2021

# Extraindo os dados

```
import requests

url = "https://dadosabertos.mec.gov.br/images/conteudo/fies/2021/relatorio_inscricao_dados_abertos_fies_22021.csv"

response = requests.get(url)
data = response.text
data
```

# Lendo os dados

```
import pyspark.pandas as ps
path_to_fies = "/FileStore/tables/aula-databricks/fies"
file_path = path_to_fies + "/fies.csv"

df_fies = ps.read_csv(file_path, sep=";")
df_fies.head()
```

# Tratamento dos dados

```
df_fies.info()
```

## Corrigindo Data Types

```
df_fies["Renda familiar mensal bruta"] = df_fies["Renda familiar mensal bruta"].str.replace(",", ".").astype(float)
df_fies["Renda mensal bruta per capita"] = df_fies["Renda mensal bruta per capita"].str.replace(",", ".").astype(float)
df_fies["Nota Corte Grupo Preferência"] = df_fies["Nota Corte Grupo Preferência"].str.replace(",", ".").astype(float)
df_fies["Matemática e suas Tecnologias"] = df_fies["Matemática e suas Tecnologias"].str.replace(",", ".").astype(float)
df_fies["Linguagens, Códigos e suas Tec"] = df_fies["Linguagens, Códigos e suas Tec"].str.replace(",", ".").astype(float)
df_fies["Ciências Humanas e suas Tec"] = df_fies["Ciências Humanas e suas Tec"].str.replace(",", ".").astype(float)
df_fies["Ciências Natureza e suas Tec"] = df_fies["Ciências Natureza e suas Tec"].str.replace(",", ".").astype(float)

df_fies.info()
```

```
df_fies["Data de Nascimento"] = ps.to_datetime(df_fies["Data de Nascimento"], format="%Y-%m-%d")
df_fies.info() 
```

## Removendo dados nulos

```
df_fies.isna().sum()
```

As colunas "Percentual de financiamento", "Semestre do financiamento" e "Qtde semestre financiado" não serão relevantes para nossa análise, portanto serão removidas.

```
30275 / df_fies.shape[0] 
```

A coluna "Renda familiar mensal bruta" possui cerca de 12% dos dados nulos. Considerando que é um dado que possa ter valores extremos, tais dados serão preenchidos com a mediana para que não haja perda de dados.

```
df_fies_cleaned = df_fies.drop(columns=["Percentual de financiamento", "Semestre do financiamento", "Qtde semestre financiado"])
df_fies_cleaned["Renda familiar mensal bruta"].fillna(df_fies_cleaned["Renda familiar mensal bruta"].median(), inplace=True)
df_fies_cleaned.info()
```

```
df_fies_cleaned.isna().sum()
```

# Qual é a correlação entre as disciplinas?

```
df_fies_cleaned.head()
```

```
df_disciplinas = df_fies_cleaned[["Redação", "Matemática e suas Tecnologias", "Linguagens, Códigos e suas Tec", "Ciências Natureza e suas Tec", "Ciências Humanas e suas Tec"]]
```

```
import plotly.graph_objects as go
corr_disciplinas = df_disciplinas.corr()

fig_corr_disciplinas = go.Figure()
fig_corr_disciplinas.add_trace(
    go.Heatmap(
        x = corr_disciplinas.columns.to_numpy(),
        y = corr_disciplinas.index.to_numpy(),
        z = corr_disciplinas.to_numpy(),
        text=corr_disciplinas.to_numpy(),
        texttemplate='%{text:.2f}'
    )
)
fig_corr_disciplinas.show()
```
